```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import math

MODEL_PATH = RUN_PATH / "weights" / "best.pt"
model = YOLO(str(MODEL_PATH))

# Carpeta donde están las imágenes prueba*.jpg
TEST_DIR = Path(r"E:\UNIVERSIDAD\PDI\Proyectobaches")

OUT_DIR = TEST_DIR / "predicciones_imagenes"
OUT_DIR.mkdir(exist_ok=True)

test_imgs = sorted(list(TEST_DIR.glob("prueba*.jpg")))

cols = 2
rows = math.ceil(len(test_imgs)/cols)

plt.figure(figsize=(14, 6 * rows))

for i, img_path in enumerate(test_imgs):
    result = model.predict(str(img_path), conf=0.40, save=False)[0]
    
    annotated = result.plot()
    annotated = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)

    # subplot
    plt.subplot(rows, cols, i+1)
    plt.imshow(annotated)
    plt.title(img_path.name)
    plt.axis("off")

    # guardar imagen
    cv2.imwrite(str(OUT_DIR / f"pred_{img_path.name}"),
                cv2.cvtColor(annotated, cv2.COLOR_RGB2BGR))

plt.tight_layout()
plt.show()

print("✅ Predicciones guardadas en:", OUT_DIR)
```

```python
import cv2
import os
from ultralytics import YOLO

# ============================
# RUTAS
# ============================
VIDEO_INPUT = r"E:\UNIVERSIDAD\PDI\Proyectobaches\Video\video_prueba.mp4"
OUT_DIR = r"E:\UNIVERSIDAD\PDI\Proyectobaches\Video\Pruebafinal"
MODEL_PATH = r"E:\UNIVERSIDAD\PDI\Proyectobaches\runs_detect\potholes_finalRTX\weights\best.pt"

# Crear carpetas de salida
os.makedirs(OUT_DIR, exist_ok=True)
frames_dir = os.path.join(OUT_DIR, "frames_annotated")
os.makedirs(frames_dir, exist_ok=True)

print("Carpeta de salida:", OUT_DIR)

# ============================
# CARGAR MODELO
# ============================
model = YOLO(MODEL_PATH)

# ============================
# ABRIR VIDEO
# ============================
cap = cv2.VideoCapture(VIDEO_INPUT)

if not cap.isOpened():
    raise SystemExit("❌ No se pudo abrir el video de entrada.")

fps = int(cap.get(cv2.CAP_PROP_FPS))
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print("Resolución:", width, "x", height, "FPS:", fps)

# Archivo de video final
output_video_path = os.path.join(OUT_DIR, "video3_detectado.mp4")
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
writer = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# ============================
# PROCESAR FRAME POR FRAME
# ============================
frame_id = 0
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

print("Procesando video... Total frames:", total_frames)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_id += 1

    # Inferencia YOLO
    res = model.predict(
        source=frame,
        conf=0.35,      # sensibilidad
        iou=0.50,
        verbose=False
    )[0]

    # Dibujar anotaciones
    annotated = res.plot()   # BGR

    # Guardar frame anotado
    frame_path = os.path.join(frames_dir, f"frame_{frame_id:05d}.jpg")
    cv2.imwrite(frame_path, annotated)

    # Escribir en video final
    writer.write(annotated)

    if frame_id % 30 == 0:
        print(f"Procesados {frame_id}/{total_frames} frames...")

cap.release()
writer.release()

print("\n====================================")
print("✔ PROCESO COMPLETADO")
print("📌 Video final guardado en:")
print(output_video_path)
print("📌 Frames guardados en:")
print(frames_dir)
print("====================================")
```

```python
from ultralytics import YOLO
import cv2
import os

# -------------------------
# Función de severidad
# -------------------------
def severidad(x1, y1, x2, y2):
    area = (x2 - x1) * (y2 - y1)
    if area > 90000:
        return "SEVERO"
    elif area > 45000:
        return "MODERADO"
    else:
        return "BAJO"

# -------------------------
# RUTAS
# -------------------------
best_weights = r"E:\UNIVERSIDAD\PDI\Proyectobaches\runs_detect\potholes_finetune\weights\best.pt"
VIDEO_IN = r"E:\UNIVERSIDAD\PDI\Proyectobaches\Video\video_prueba.mp4"
OUT_DIR = r"E:\UNIVERSIDAD\PDI\Proyectobaches\Video\Pruebafinal"

os.makedirs(OUT_DIR, exist_ok=True)
VIDEO_OUT = os.path.join(OUT_DIR, "video3_detectado_severidad.mp4")

# -------------------------
# PROCESAMIENTO DE VIDEO
# -------------------------
model = YOLO(best_weights)
cap = cv2.VideoCapture(VIDEO_IN)

fps = cap.get(cv2.CAP_PROP_FPS)
w   = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h   = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(VIDEO_OUT, fourcc, fps, (w, h))

print("[INFO] Procesando video...")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(frame, conf=0.20, iou=0.4, imgsz=960)
    res = results[0]

    annotated = res.plot()  # BGR

    # Agregar severidad
    for b in res.boxes:
        x1, y1, x2, y2 = map(int, b.xyxy[0])
        sev = severidad(x1, y1, x2, y2)
        cv2.putText(
            annotated,
            sev,
            (x1, y1 - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 0, 255),
            2
        )

    out.write(annotated)

cap.release()
out.release()
print("[OK] Video exportado en:", VIDEO_OUT)
```
